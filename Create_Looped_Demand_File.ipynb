```python
import pandas as pd
import numpy as np
from datetime import datetime, date, time, timedelta
```

```python
def keep_exactly_4_years(df, start_year, start_month, start_day, start_hour):
    to_remove = []
    start_dt = datetime.strptime(f'{start_year}{start_month}{start_day}{start_hour}', '%Y%m%d%H')
    end_dt = datetime.strptime(f'{start_year+4}{start_month}{start_day}{start_hour}', '%Y%m%d%H')
    print(f"Initial length {len(df.index)}")
    print(start_dt, end_dt)
    
    for idx in range(len(df.index)):
        
        if idx == 0 or idx == len(df.index)-1:
            print(df.loc[idx, 'year'], df.loc[idx, 'month'], df.loc[idx, 'day'])
    
        d = date(df.loc[idx, 'year'], df.loc[idx, 'month'], df.loc[idx, 'day'])
        t = time(df.loc[idx, 'hour']-1, 0) # silly SEM hour formatting
        current_time = datetime.combine(d, t)
        
        if current_time < start_dt:
            to_remove.append(idx)
        elif current_time >= end_dt:
            to_remove.append(idx)
    
    print(f"length to remove {len(to_remove)}")
    df = df.drop(index=to_remove)

    print(f"Final length {len(df.index)}")
    print(f"First entry: {df.iloc[0]['year']} {df.iloc[0]['month']} {df.iloc[0]['day']}")
    print(f"Last  entry: {df.iloc[len(df.index)-1]['year']} {df.iloc[len(df.index)-1]['month']} {df.iloc[len(df.index)-1]['day']}")
        
    return df

def incriment_years(df, X):
    if X == 0:
          return df
    df['year'] = np.where(df['year'] == 2015, 2015-X, df['year'])
    df['year'] = np.where(df['year'] == 2015+1, 2015-X+1, df['year'])
    df['year'] = np.where(df['year'] == 2015+2, 2015-X+2, df['year'])
    df['year'] = np.where(df['year'] == 2015+3, 2015-X+3, df['year'])
    df['year'] = np.where(df['year'] == 2015+4, 2015-X+4, df['year'])
    return df

          
          
def check_uniqueness(df):
    

    d = date(df.iloc[0]['year'], df.iloc[0]['month'], df.iloc[0]['day'])
    t = time(df.iloc[0]['hour']-1, 0) # silly SEM hour formatting
    prev_dt = datetime.combine(d, t)
          
    for idx in range(1, len(df.index)):
        
        d = date(df.iloc[idx]['year'], df.iloc[idx]['month'], df.iloc[idx]['day'])
        t = time(df.iloc[idx]['hour']-1, 0) # silly SEM hour formatting
        current_time = datetime.combine(d, t)

        if not prev_dt + timedelta(hours=1) == current_time:
            print(idx, df.iloc[idx]['year'], df.iloc[idx]['month'], df.iloc[idx]['day'], df.iloc[idx]['hour'])
        
        prev_dt = current_time
```
